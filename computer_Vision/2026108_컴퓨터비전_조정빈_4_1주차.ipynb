```python
def show_image(title,img):
  rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.figure(figsize=(10,10))
  plt.imshow(rgb_img)
  plt.title(title)
  plt.imshow(rgb_img)
  plt.show()
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

image_line = np.full((512, 512, 3), 255, dtype=np.uint8)
image_line = cv2.line(
    image_line,
    (0,0),
    (255, 255),
    (255, 0, 0),
    3
)

show_image("Line Drawing", image_line)
```

```python
image_rectangle = np.full((512, 512, 3), 255, dtype=np.uint8)
image_rectangle = cv2.rectangle(
    image_rectangle,
    (0,0), #왼쪽 상단 꼭짓점
    (255, 255), #오른쪽 하단 꼭짓점
    (255, 0, 0), # 파란색
    3 #선두께
)

show_image("Rectangle Drawing", image_rectangle)
```

```python
imgage_circle = np.full((512, 512, 3), 255, dtype=np.uint8)
imgage_circle = cv2.circle(
    imgage_circle,
    (256, 256), #원의 중심
    80, #반지름
    (255, 0, 0), #파란색
    3 #선두께
)

show_image("Circle Drawing", imgage_circle)
```

```python
image_poly = np.full((512, 512, 3), 255, dtype=np.uint8)

points = np.array([[50, 50], [150, 200], [300, 100], [400, 400]], np.int32)
points = points.reshape((-1, 1, 2))

image_poly = cv2.polylines(
    image_poly,
    [points], #리스트형태로
    True, #닫힌도형
    (255, 0, 0), #빨간색
    3
)

show_image("Polygon Drawing", image_poly)
```

```python
image_ellipse = np.full((512, 512, 3), 255, dtype=np.uint8)
image_ellipse = cv2.ellipse(
    image_ellipse,
    (256, 256), #원의 중심
    (100, 50), #반지름
    30,
    0, 360,
    (0, 255, 0), #초록색
    3
)

show_image("Ellipse Drawing", image_ellipse)
```

```python
image_text = np.full((512, 512, 3), 255, np.uint8)
image_text = cv2.putText(
    image_text,
    'Hello World',
    (50, 200),
    cv2.FONT_HERSHEY_SIMPLEX,
    2,
    (0, 0, 255),
)

show_image("Text Drawing", image_text)
```

```python
image_all = np.full((512, 512, 3), 255, np.uint8)

cv2.line(image_all, (0,0), (255, 255), (255, 0, 0), 3)
cv2.rectangle(image_all, (0,0), (255, 255), (255, 0, 0), 3)
cv2.circle(image_all, (256, 256), 80, (255, 0, 0), 3)
cv2.ellipse(image_all, (256, 256), (100, 50), 30, 0, 360, (0, 255, 0), 3)
cv2.putText(image_all, 'Hello World', (50, 200), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255),)

show_image("All Drawing", image_all)
```

```python
#애니메이션 효과(for 루프를 돌면서 원의 위치나 반지름 변경해 매 반복마다 이미지 업데이트)
import time
import cv2
import numpy as np

image = np.zeros((512, 512, 3), np.uint8)

x, y = 256, 256
radius = 50
dx, dy = 5, 5
dr = 2

for i in range(100):
    image.fill(0)

    cv2.circle(image, (x, y), radius, (0, 0, 255), -1)

    x += dx
    y += dy
    radius += dr

    if x + radius > 512 or x - radius < 0:
        dx = -dx
    if y + radius > 512 or y - radius < 0:
        dy = -dy
    if radius > 100 or radius < 20:
        dr = -dr

    from IPython.display import display
    import matplotlib.pyplot as plt

    def show_image(title,img):
        rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(10,10))
        plt.imshow(rgb_img)
        plt.title(title)
        plt.imshow(rgb_img)
        plt.show()

    show_image("Animation", image)

    time.sleep(0.05)
```

```python
#ROI 내부만 도형 그리기 (100x100 사각형만 잘라내 그 부분에만 선, 원, 텍스트 등 추가)
import cv2
import numpy as np

image = np.zeros((512, 512, 3), np.uint8)

x, y, w, h = 100, 100, 100, 100

roi = image[y:y+h, x:x+w]

cv2.line(roi, (0, 0), (w, h), (0, 255, 0), 2)
cv2.circle(roi, (w//2, h//2), 20, (0, 0, 255), -1)
cv2.putText(roi, 'Hello', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
```
